In Part I, data is collected using the API, and cleaned in Part II. In Part III, visualization is performed to detect any possible outliers, anomalies, and build intuition around the data before performing machine learning modelling in Part IV.

### Step 1: Import the following libraries
- pandas
- matplotlib.pyplot as plt
- seaborn as sns

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Read the CSV from Part II
Read the CSV from Part II using pandas.

Things to take note of: 
1. make sure 'time' is index
2. parse the datetime in 'time' so that it's a proper DateTime object

```python
# Step 2: Read CSV as DateTime and make sure the dates are parsed
data = pd.read_csv('Ordered_DataFrame.csv', parse_dates=True)
data['time'] = pd.to_datetime(data['time'])
data = data.set_index('time')
display(data)
```

```python
data = data.iloc[: , 1:]
data
```

### Step 3: Plot 'properties.taxi_count'
Onwards to the first plot! 

```python
# Step 3: Plot properties.taxi_count vs time
plt.figure(figsize = (20, 8))
plt.plot(data.index.values, data['properties.taxi_count'])
plt.title("'properties.taxi_count' vs. Time", fontsize = 14, fontweight = "bold")
plt.xlabel("Time")
plt.ylabel("properties.taxi_count")
plt.show()
```

### Step 4: Investigate the odd datapoints
The visualization suggests that there are 0 taxis available in Singapore on certain days. This simply cannot be. 

Before proceeding further with visualization, let's either remove or fix this problem. 

Check the parts in the DataFrame where "properties.taxi_count" == 0. Slice the DataFrame based on the column values.

```python
# Step 4: Get the slice where there is no taxi counts
no_taxi = data.loc[data['properties.taxi_count'] == 0]
no_taxi
```

### Step 5: Remove those erroneous rows
Looks like there was some downtime on those days, since all of the columns contain 0. As such, remove them. This is where visualization helps in identifying anomalous parts of the dataset. 

```python
# Step 5: Remove rows where properties.taxi_count is 0
has_taxi = data[data['properties.taxi_count'] != 0]
display(has_taxi)
```

### Step 6: Resample the DataFrame by day
The current plot seems very chaotic, and it's going to be hard to intuit anything from it, e.g., whether the data is cyclical, etc. 

As such, let's resample the DataFrame and take the median! This is so that intra-day variation can be reduced and focus on the median to identify if there is any possible pattern.

In a median resample, the DataFrame index now considers only the <strong>Day</strong>. Notice how the hours have disappeared, leaving only the date. 

```python
# Step 6: Resample DataFrame by day, and take the day's median value
has_taxi_daily = has_taxi.resample('D').median()
has_taxi_daily
```

### Step 7: Plot 'properties.taxi_count' from the resampled DataFrame over original plot
After resampled the DataFrame by day, it's time to plot it to see the effects of analysis. 

```python
# Step 7: Plot resampled DataFrame over original plot
labels = ["Intraday", "Daily"]

plt.figure(figsize = (20, 8))
plt.plot(data.index.values, data['properties.taxi_count'])
plt.plot(has_taxi_daily.index.values, has_taxi_daily['properties.taxi_count'])
plt.title("'properties.taxi_count' vs. Time", fontsize = 14, fontweight = "bold")
plt.xlabel("Time")
plt.ylabel("properties.taxi_count")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

Once resampled, there is less intraday variation since only the median of the day is taken. More importantly, the rough cyclic nature of the available taxis throughout the month is visible.

### Step 10: Get the hourly average of taxi availability
Get the hourly average data of the DataFrame. This means averaging all of the taxi availability for the hours of the day, across all days. 

Compared to what is done in Step 9 where it was a resampling, what is done here is to group the 'properties.taxi_count' by the hour and followed by the averaging. 

```python
# Step 10: Get the DataFrame containing the hourly average
has_taxi_hourly = has_taxi.groupby(has_taxi.index.hour).mean()
has_taxi_hourly
```

### Step 11: Plot the hourly average of 'properties.taxi_count' using barplot
Time to plot the the averages to see if there's a pattern.

```python
# Step 11: Plot the hourly average of 'properties.taxi_count'
plt.figure(figsize = (9, 6))
plt.bar(x = has_taxi_hourly.index.values, height = has_taxi_hourly['properties.taxi_count'], color = "midnightblue")
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Hourly Average of 'properties.taxi_count'", fontsize = 16, fontweight = "bold")
plt.xlabel("Time by Hour", fontsize = 13)
plt.ylabel("properties.taxi_count", fontsize = 13)
plt.show()
```
